```python
import pickle

with open('urls_comedy_covid_644.pkl', 'rb') as picklefile:
    urls = pickle.load(picklefile)
```

```python
from pymongo import MongoClient

client = MongoClient()
db = client['comedy_covid']
db.create_collection("covid_comedy_youtube_transcripts2")
covid_comedy_youtube_transcripts2 = db.get_collection('covid_comedy_youtube_transcripts2')
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

for url in urls:
    try:
        url = url.replace("https://www.youtube.com/watch?v=", '')
        response = YouTubeTranscriptApi.get_transcript(url)
        covid_comedy_youtube_transcripts2.insert_one({'url': url, 'transcript': response})
    except:
        covid_comedy_youtube_transcripts2.insert_one({'url': "--", 'transcript': "--"})
```
